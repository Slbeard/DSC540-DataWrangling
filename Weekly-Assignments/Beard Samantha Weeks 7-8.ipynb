### CH 7 Method 1: Manipulate Strings

```python
# Remove | from headers
candy_df.columns = candy_df.columns.str.replace("|", ":", regex=True)

# Remove Q#: from headers
candy_df.columns = candy_df.columns.str.split(': ', n=1).str.get(-1)

print(candy_df.head(5))
```

### CH 7 Method 2: Replace values

```python
# replace joy with love
candy_df = candy_df.replace(['JOY'], 'LOVE')
candy_df.head(5)
```

### CH 11 Method 1: Convert between string and date time

```python
#assuming that the data was taken from the last week, I will convert Friday to 2/9 and Sunday to 2/11

#first I have to change the string to the date
candy_df['DAY'] = candy_df['DAY'].str.replace(r"Friday","02/09/2024 12:00:00", regex=True)
candy_df['DAY'] = candy_df['DAY'].str.replace(r"Sunday","02/11/2024 12:00:00", regex=True)
```

```python
# then I can change from string to datetime
candy_df['DAY'] = pd.to_datetime(candy_df['DAY'])
```

```python
pd.crosstab(candy_df['DAY'], candy_df['GENDER'])
```

### CH 11 Method 2: Convert timestamps to periods and back

```python
candy_df['DAY'] = candy_df['DAY'].dt.to_period('M')
```

```python
pd.crosstab(candy_df['DAY'], candy_df['GENDER'])
```

```python
candy_df['DAY'] = candy_df['DAY'].dt.to_timestamp()
```
