```python
# import packages

import pandas as pd
import re
import numpy as np
```

```python
# import cvs file to dataframe
grammy_awards_df = pd.read_csv('the_grammy_awards.csv')

#show top 5
grammy_awards_df.head()
```

```python
# row count
startingRowCount = len(grammy_awards_df.index)
startingRowCount
```

```python
num_cols = grammy_awards_df.shape
print(num_cols[1])
```

```python
# remove words within parentheses using pandas.series.str.replace to replace each occurrence of () with empty string.
grammy_awards_df['title'] = grammy_awards_df['title'].str.replace(r"\(.*\)","", regex=True)
grammy_awards_df.head()
```

```python
grammy_awards_df = grammy_awards_df.drop(columns = ['updated_at'])
grammy_awards_df.head()
```

```python
grammy_awards_df.category.unique()
```

```python
grammy_awards_df = grammy_awards_df.dropna()
```

```python
print(grammy_awards_df['artist'].sample(50))
```

```python
grammy_awards_df['artist'] = grammy_awards_df['artist'].str.split(',').str[0]
```

```python
grammy_awards_df['artist'] = grammy_awards_df['artist'].str.split('&').str[0]
```

```python
grammy_awards_df['artist'] = grammy_awards_df['artist'].str.split('(').str[0]
```

```python
grammy_awards_df['artist'] = grammy_awards_df['artist'].str.split('Featuring').str[0]
```

```python
grammy_awards_df['title'] = grammy_awards_df['title'].str.split(' ').str[0]
```

```python
grammy_awards_df['artist'].fillna("Various", inplace = True)
```

```python
grammy_awards_df['artist'] = grammy_awards_df['artist'].str.title()
```

```python
grammy_awards_df['nominee'] = grammy_awards_df['nominee'].str.title()
```

```python
artists_list = grammy_awards_df.artist.unique()
print(len(artists_list))
```
